# Practice 3 Q3b - Vander Pol Oscillator

## $\alpha < 0$

```python
# set parameters
alpha = -1.

def VO_model(t,x):
    xout = VanderPol(t,x,alpha)
    return xout
```

```python
# set initial conditions
x0 = np.array([0.1,0.1]);
```

```python
# IVP solver settings
tspan = [0,100]
h = 0.01

xout,t,xend = MyIVP(VO_model,x0,tspan,h)

xout = np.squeeze(xout)
```

### Plot solution in phase plane

```python
fig, axs = plt.subplots(1, 2, layout='constrained', figsize=(10, 5))

ax1 = axs[0]
ax1.plot(xout[0,:],xout[1,:])
ax1.set_xlabel("$x$")
ax1.set_ylabel("$\dot{x}$")
ax1.set_title("Solution Trajectory in Phase Plane")
ax1.set_xlim([-0.2, 0.2])
ax1.set_ylim([-0.2, 0.2])

ax2 = axs[1] 
ax2.plot(t,xout[0,:])
ax2.plot(t,xout[1,:])
ax2.set_xlabel("$t$")
ax2.set_title("Solution Trajectories in time")
ax2.legend(['$x$',"$\dot{x}$"])
ax2.set_xlim([0, 100])

plt.show()
```
