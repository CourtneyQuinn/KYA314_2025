## Part d

```python
# set parameters
alpha = 1.4
beta = 0.25

p = np.array([alpha,beta])

# set number of iterations
Nsteps = 100

# set initial conditions
x0 = np.array([[1,-1],[1,-1]])
```

```python
# create empty solution vector 
xtraj1 = np.empty((x0.shape[0],Nsteps+1))
xtraj1[:] = np.nan
xtraj2 = np.empty((x0.shape[0],Nsteps+1))
xtraj2[:] = np.nan

# input initial condition
xtraj1[:,0] = x0[:,0]
xtraj2[:,0] = x0[:,1]
```

### Iterate map

```python
for i in np.arange(0,Nsteps):
    x0 = DuffingMap(x0,p)
    xtraj1[:,i+1] = x0[:,0]
    xtraj2[:,i+1] = x0[:,1]
```

### Plot solution

```python
fig, axs = plt.subplots(1, 2, layout='constrained', figsize=(10, 5))

ax1 = axs[0]
ax1.plot(xtraj1[0,50:],xtraj1[1,50:],'.')
ax1.plot(xtraj2[0,50:],xtraj2[1,50:],'.')
ax1.set_xlabel("$x_n$")
ax1.set_ylabel("$y_n$")
ax1.set_title("Solution Trajectories in Phase Plane")
#ax1.set_xlim([-2, 2])
#ax1.set_ylim([-0.5, 0.5])

ax2 = axs[1] 
ax2.plot(np.arange(0,Nsteps+1),xtraj1[0,:],'.')
ax2.plot(np.arange(0,Nsteps+1),xtraj2[0,:],'.')
ax2.set_xlabel("$n$")
ax2.set_ylabel("$x_n$")
ax2.set_title("Solution Trajectories ($x_n$)")
ax2.set_xlim([0, 100])

plt.show()
```

### Interpretation
We can see from the phase plane plot above that the solutions converge to stable fixed points. Interestingle they cross to the opposite part of phase space. This is further suported by right figure, where we see convergence by 20 iterations.
