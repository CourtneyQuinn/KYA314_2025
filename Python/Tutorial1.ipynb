# Set-up iterations

````python
# define parameters
lam = 1.8
x0 = np.array([0.1])
Nsteps = 1000
````

````python
# create empty solution vector 
xtraj = np.empty((Nsteps+1,1));
xtraj[:] = np.nan

# input initial condition
xtraj[0,:] = x0;
````

# Iterate map

````python
for i in np.arange(0,Nsteps):
    x0 = LogisticMap(x0,lam);
    xtraj[i+1,:] = x0;
````

# Plot solutions

````python
fig = plt.figure(figsize=(5,5))
axs = plt.plot(xtraj[100:-1,:],xtraj[101:,:],'.',c='k')
plt.xlabel("$x_n$")
plt.ylabel("$x_{n+1}$")
plt.title("Solution Trajectory in Phase Plane")
plt.xlim([-np.abs(lam)-1, np.abs(lam)+1])
plt.ylim([-np.abs(lam)-1, np.abs(lam)+1])
plt.show()
````

````python
fig = plt.figure(figsize=(5,5))
axs = plt.plot(np.arange(0,Nsteps+1),xtraj,'r')
plt.xlabel("$n$")
plt.ylabel("$x_n$")
plt.title("Solution Trajectory")
plt.xlim([0, 1000])
plt.show()
````
