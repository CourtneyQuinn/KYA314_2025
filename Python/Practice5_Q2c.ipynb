# Practice 5 Q2c - Homoclinic orbits

## Hamiltonian

The Hamiltonian is 
$$ H(\theta, v) = -\frac{v^2}{2}+\alpha^2\cos(\theta) $$
At saddle, $H=\alpha^2$ 

We can rearrange for $v$ to give the curve for the Hamiltonian through the saddles: 
$$v(\theta) = \pm\sqrt{2\alpha^2\cos(\theta)+2\alpha^2}.$$

```python
# define function for Hamiltonian curve
v= lambda theta,alpha: np.sqrt(2*alpha**2*np.cos(theta)+2*alpha**2)
```

## Calculate Hamiltonian curve when varying $\alpha$

```python
alpha1 = 0.5
alpha2 = 1
alpha3 = 2

thetas = np.linspace(-np.pi,np.pi,628)
v1 = v(thetas,alpha1)
v2 = v(thetas,alpha2)
v3 = v(thetas,alpha3)
```

### Plot curves in phase plane

```python
plt.plot(thetas,v1)
plt.plot(thetas,v2)
plt.plot(thetas,v3)
plt.plot(thetas,-v1, '#1f77b4')
plt.plot(thetas,-v2, '#ff7f0e')
plt.plot(thetas,-v3, '#2ca02c')
plt.xlabel("$\\theta$")
plt.ylabel("$v$")
plt.title("Hamiltonian curves in Phase Plane")
#plt.xlim([-0.1, 1])
#plt.ylim([-0.1, 5])
plt.legend(['$\\alpha=0.5$','$\\alpha=1$','$\\alpha=2$'])

plt.show()
```
