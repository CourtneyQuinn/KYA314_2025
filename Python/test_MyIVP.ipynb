# Test Newton solver using provided example

## Example
$\ddot{x}=-x-\dot{x}+\cos(t)$

Turn into 2D ODE system

$ \begin{align} \dot{x} =& v \\ \dot{v} =& -x-v+\cos(t) \end{align}$

Solution is
$\dot{x}=\sin(t)$ and $x=\cos(t)$

```python
# Define function
def f(t,x):
    f1 = x[1,:]
    f2 = -x[0,:]-x[1,:]+np.cos(t)
    fout = np.array([f1,f2])
    return fout
```

### Define IVP inputs

```python
tspan = [0,2*np.pi];
h=0.05;
x0 = np.array([[0.],[1.]]);
```

```python
# check function with initial condition
f0 = f(0.,x0)
print(f0)
```

### Solve IVP

```python
[xt,t,xend]=MyIVP(f,x0,tspan,h)
```

## Plot solution

```python
plt.plot(t,xt[0,:].T)
plt.plot(t,xt[1,:].T)
plt.xlabel("$t$")
plt.title("Solution Trajectory in Time")
plt.xlim([0, 2*np.pi])
plt.legend(["v","x"])
plt.show()
```
