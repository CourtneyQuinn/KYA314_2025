## Define the system

```python
def Rossler(t,x,p):
    f1 = -x[1,:]-x[2,:]
    f2 = x[0,:]+p[0,:]*x[1,:]
    f3 = p[1,:]+x[2,:]*(x[0,:]-p[2,:])
    fout = np.array([f1,f2,f3])
    return fout
```

## Set parameters

```python
b = 0.1
c = 6

a1 = -3 # stable equilibium
a2 = 0.01 # stable periodic orbit
a3 = 0.2 # chaotic attractor

ps=np.array([[a1,a2,a3],[b,b,b],[c,c,c]])
```

## Simulate behaviour

```python
tspan = [0,2000]
h = 1e-2
x0s = np.ones((3,3))*0.1

xout,t,xend = MyIVP(lambda t,x:Rossler(t,x,ps),x0s,tspan,h)
```

## Plot solutions

```python
fig = plt.figure(figsize=(15,5))

# Phase plane
ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.plot(xout[0,0,-50000:],xout[1,0,-50000:],xout[2,0,-50000:],'.')
ax1.set_xlabel("$x$")
ax1.set_ylabel("$y$")
ax1.set_zlabel("$z$")
ax1.set_title("Solution Trajectory in Phase Plane")

# Trajectory in time
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(t[-50000:],xout[:,0,-50000:].T)
ax2.set_xlabel("$t$")
ax2.legend(['$x$','$y$','$z$'])
ax2.set_title("Solution Trajectory")

plt.show()
```
