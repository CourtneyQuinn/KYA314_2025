## Define the system

```python
f = lambda t,x,p :np.array([x[1,:],
    -p[0]*x[0,:]-p[1]*x[0,:]**3+p[2]*np.cos(t)])

alpha=-1.
beta=0.25
```

## No forcing

```python
p=np.array([alpha,beta,0.])

tspan = [0,100]
h = 1e-2
x0 = np.array([1e-3,1e-3])

xout,t,xend = MyIVP(lambda t,x:f(t,x,p),x0,tspan,h)

fig, axs = plt.subplots(1, 2, layout='constrained', figsize=(10, 5))

ax1 = axs[0]
ax1.plot(xout[0,0,:],xout[1,0,:])
ax1.set_xlabel("$x$")
ax1.set_ylabel("$\dot{x}$")
ax1.set_title("Solution Trajectory in Phase Plane")

ax2 = axs[1] 
ax2.plot(t,np.squeeze(xout).T)
ax2.set_xlabel("$t$")
ax2.legend(['$x$','$\dot{x}$'])
ax2.set_title("Solution Trajectory")

plt.show()
```

## Weak forcing

```python
p=np.array([alpha,beta,0.01])

tspan = [0,100]
h = 1e-2
x0 = np.array([1e-3,1e-3])

xout,t,xend = MyIVP(lambda t,x:f(t,x,p),x0,tspan,h)

fig, axs = plt.subplots(1, 2, layout='constrained', figsize=(10, 5))

ax1 = axs[0]
ax1.plot(xout[0,0,:],xout[1,0,:])
ax1.set_xlabel("$x$")
ax1.set_ylabel("$\dot{x}$")
ax1.set_title("Solution Trajectory in Phase Plane")

ax2 = axs[1] 
ax2.plot(t,np.squeeze(xout).T)
ax2.set_xlabel("$t$")
ax2.legend(['$x$','$\dot{x}$'])
ax2.set_title("Solution Trajectory")

plt.show()
```
